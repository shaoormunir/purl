```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
from tqdm.auto import tqdm
tqdm.pandas()
```

```python
RESULTS_DIR = 'results/20k-run-6-19-23'
```

```python
df_int = pd.DataFrame()

with open(os.path.join(RESULTS_DIR, 'unlabelled/interpretations'), 'r') as f:
    data = json.load(f)
```

```python
# read file line by line
data_unlabelled = []
with open(os.path.join(RESULTS_DIR, 'unlabelled/predictions'), 'r') as f:
    for line in tqdm(f):
      data = line.split('|$|')
      # remove whitespace
      data = [x.strip() for x in data]
      # remove empty strings
      data = [x for x in data if x]

      data_unlabelled.append(data)
      # store in dataframe
      # df_unlabeled.loc[len(df_unlabeled)] = data
```

```python
df_unlabelled = pd.DataFrame(data_unlabelled, columns=['prediction', 'name', 'visit_id'])
```

```python
dict_data = []
for key in tqdm(data):
  temp = {}
  temp['name'] = data[key]['name']
  temp['visit_id'] = data[key]['vid']

  for contribution in data[key]['contributions']:
    temp[contribution[1]] = contribution[0]
  
  dict_data.append(temp)
```

```python
df_int = pd.DataFrame(dict_data)
```
